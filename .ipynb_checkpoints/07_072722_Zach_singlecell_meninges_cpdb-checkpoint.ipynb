```r
# ok let's read in the cpdb output files.
naive_pvals <- read.delim("naive_cpdb/pvalues.txt", check.names = FALSE)
naive_means <- read.delim("naive_cpdb/means.txt", check.names = FALSE)
naive_decon <- read.delim("naive_cpdb/deconvoluted.txt", check.names = FALSE)
```

```r
vsv_pvals <- read.delim("vsv_cpdb/pvalues.txt", check.names = FALSE)
vsv_means <- read.delim("vsv_cpdb/means.txt", check.names = FALSE)
vsv_decon <- read.delim("vsv_cpdb/deconvoluted.txt", check.names = FALSE)
```

```r
library(dplyr)
```

```r
# merge the output
pvals <- full_join(naive_pvals, vsv_pvals, by = c(colnames(naive_pvals)[1:11]))
pvals[is.na(pvals)] <- 1 # convert missing to 1
```

```r
means <- full_join(naive_means, vsv_means, by = c(colnames(naive_means)[1:11]))
means[is.na(means)] <- 0 # convert missing to 0
```

```r
decon <- full_join(naive_decon, vsv_decon, by = c(colnames(naive_decon)[1:6]))
decon[is.na(decon)] <- 0 # convert missing to 0
```

```r
options(repr.plot.height = 3.5, repr.plot.width = 25)
plot_cpdb(cell_type1 = 'B_GC', cell_type2 = '.',
          scdata = sce,
          idents = 'celltype', # column name where the cell ids are located in the metadata
          split.by = 'condition', # column name where the grouping column is. Optional.
          means = means,
          pvals = pvals,
          keep_significant_only = FALSE,
          standard_scale = TRUE,
          genes = c('CXCL13'))
```

```r
options(repr.plot.height = 3.5, repr.plot.width = 5)
p <- plot_cpdb(cell_type1 = 'B_GC', cell_type2 = 'Tfh',
          scdata = sce,
          idents = 'celltype', # column name where the cell ids are located in the metadata
          split.by = 'condition', # column name where the grouping column is. Optional.
          means = means,
          pvals = pvals,
          keep_significant_only = TRUE,
          standard_scale = TRUE,
          genes = c('IL21 receptor', 'ICOSLG', 'CD40', 'CD28', 'CTLA4', 'IL10'),
          highlight_size = 1
) + guides(fill=guide_legend(ncol=1), colour = FALSE) + theme(legend.box = "horizontal") 
p$data$Var1 <- factor(p$data$Var1, levels = rev(c('CD28-CD80', 'CD28-CD86', 'CD40-CD40LG', 'IL21 receptor-IL21', 'IL10 receptor-IL10', 'CTLA4-CD80', 'CTLA4-CD86')))
p$data$Var2 <- factor(p$data$Var2, levels = c('naive_B_GC-naive_Tfh',
                                              'VSV_B_GC-VSV_Tfh',
                                              'naive_Tfh-naive_B_GC',
                                              'VSV_Tfh-VSV_B_GC'))
p
ggsave('plot_cpdb.pdf', plot = p, h = 3, w = 1, useDingbats = FALSE)
```

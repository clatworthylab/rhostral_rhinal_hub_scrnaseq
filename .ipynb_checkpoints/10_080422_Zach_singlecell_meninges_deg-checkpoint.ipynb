```python
## make a combined TFH + B cell object just to plot genes
# re-compute umap for B cells with Tfr and Tfh
bdata = adata[adata.obs['celltype'].isin(['B-T_multiplets', 'B_GC', 'B_developing', 'B_Plasma', 'Tfh', 'Tfr'])].copy()
bdata = bdata.raw.to_adata()
sc.pp.highly_variable_genes(bdata, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(bdata)
```

```python
import re
for i in bdata.var.index:
    if re.search('^Igh', i):
        bdata.var.at[i, 'highly_variable'] = False
```

```python
bdata.raw = bdata
bdata = bdata[:, bdata.var.highly_variable].copy()
bdata
```

```python
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt', 'S_score', 'G2M_score'])
# sc.pp.regress_out(bdata, ['total_counts', 'pct_counts_mt'])
sc.pp.scale(bdata, max_value=10)
```

```python
sc.tl.pca(bdata, svd_solver='arpack', n_comps=100)
sc.pl.pca_variance_ratio(bdata, log=True, n_pcs=100)
```

```python
sc.external.pp.harmony_integrate(bdata, key = 'sequencing_batch', adjusted_basis='X_pca', max_iter_harmony = 30)
```

```python
sc.pp.neighbors(bdata, n_neighbors=10, n_pcs = 60)
sc.tl.umap(bdata, min_dist=.3)
```

```python
sc.pl.umap(bdata, color=['celltype'], size = 30)
```

```python
bdata.obs['celltype_condition'] = [a + '_' + b for a, b in zip(bdata.obs['celltype'], bdata.obs['condition'])]
bdata.obs['celltype_condition'] = bdata.obs['celltype_condition'].astype('category')
catorder_a = bdata.obs['celltype'].cat.categories
catorder_b = bdata.obs['condition'].cat.categories
new_order = []
for x in catorder_a:
    for y in catorder_b:
        new_order.append(x + '_' + y)
bdata.obs['celltype_condition'] = bdata.obs['celltype_condition'].cat.reorder_categories(new_order)
```

```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import os
import tools

sc.logging.print_header()

os.chdir(os.path.expanduser('/home/jovyan/Zach/scanpy/'))
adata = sc.read_h5ad('h5ad/meninges_integrated.h5ad')
adata
```

```python
sc.pl.umap(adata, color = 'celltype')
```

```python
# make a special exception to keep the B-T multiples, but all other multiplets are to be removed.
adata.obs['keep'] = 'True'
for i in adata.obs.index:
    if adata.obs.loc[i, 'celltype'] != 'B-T_multiplets':
        if adata.obs.loc[i, 'celltype'] == 'Multiplets':
            adata.obs.at[i, 'keep'] = 'False'
        if adata.obs.loc[i, 'is_doublet'] == 'True':
            adata.obs.at[i, 'keep'] = 'False'
```

```python
sc.pl.umap(adata, color = 'keep')
```

```python
adata = adata[adata.obs['keep'] == 'True'].copy()
adata = adata.raw.to_adata()
adata
```

```python
sc.pp.highly_variable_genes(adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(adata)
```

```python
adata.raw = adata
```

```python
adata = adata[:, adata.var.highly_variable].copy()
adata
```

```python
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt', 'S_score', 'G2M_score'])
sc.pp.scale(adata, max_value=10)
```

```python
sc.tl.pca(adata, svd_solver='arpack', n_comps=100)
sc.pl.pca_variance_ratio(adata, log=True, n_pcs=100)
```

```python
sc.external.pp.harmony_integrate(adata, key = 'sequencing_batch', adjusted_basis='X_pca')
```

```python
sc.pl.pca_variance_ratio(adata, log=True, n_pcs=100)
```

```python
sc.pp.neighbors(adata, n_neighbors=10, n_pcs = 60)
sc.tl.leiden(adata)
sc.tl.umap(adata, min_dist=.3)
```

```python
sc.pl.umap(adata, color=['sample_id', 'leiden'])
```

```python
sc.pl.umap(adata, color=['celltype'])
```

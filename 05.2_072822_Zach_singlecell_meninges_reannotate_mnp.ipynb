```python
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import os
import tools

sc.logging.print_header()

os.chdir(os.path.expanduser('/home/jovyan/Zach/scanpy/'))
adata = sc.read_h5ad('h5ad/meninges_integrated.h5ad')
adata
```

```python
mdata = adata[adata.obs['celltype'].isin(['Monocyte', 'Macrophage'])].copy()
mdata
```

```python
mdata = mdata.raw.to_adata()
mdata
```

```python
sc.pp.highly_variable_genes(mdata, min_mean=0.0125, max_mean=3, min_disp=0.5)
sc.pl.highly_variable_genes(mdata)
```

```python
mdata.raw = mdata
```

```python
mdata = mdata[:, mdata.var.highly_variable].copy()
mdata
```

```python
sc.pp.regress_out(mdata, ['total_counts', 'pct_counts_mt'])
sc.pp.scale(mdata, max_value=10)
```

```python
sc.tl.pca(mdata, svd_solver='arpack', n_comps=50)
sc.pl.pca_variance_ratio(mdata, log=True, n_pcs=50)
```

```python
sc.external.pp.harmony_integrate(mdata, key = 'sequencing_batch', adjusted_basis='X_pca', max_iter_harmony = 30)
```

```python
sc.pl.pca_variance_ratio(mdata, log=True, n_pcs=50)
```

```python
sc.pp.neighbors(mdata, n_neighbors=10, n_pcs = 20)
sc.tl.leiden(mdata)
sc.tl.umap(mdata, min_dist=.3)
```

```python
sc.pl.umap(mdata, color=['sample_id', 'leiden'])
```

```python
sc.tl.leiden(mdata, restrict_to = ('leiden', ['11']), resolution = .3)
sc.pl.umap(mdata, color = 'leiden_R')
```
